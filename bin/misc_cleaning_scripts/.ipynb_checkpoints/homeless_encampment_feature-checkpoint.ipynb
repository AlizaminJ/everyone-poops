```python
import pandas as pd
import numpy as np
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon

#empty_matrix = 
encampments = pd.read_csv('..//..//data//sf_open//311_Cases.csv')
blocks = pd.read_csv('..//..//data//census//tl_2010_06075_tabblock10.csv')
```

```python
encampments = encampments[encampments['Request Type']=='Encampment Reports']
```

```python
# get the hour
encampments['hour'] = encampments['Opened'].apply(lambda x:int(x[11:13]))

def bin_hours(row):
    """
    will group hours together in order to have bins instead of the raw timestamp
    """
    if row['hour'] < 6:
        return '0' 
    elif row['hour'] < 12:
        return '6' 
    elif row['hour'] < 18:
        return '12'
    else:
        return '18'

# insert the rounded hour back into the dataframe
encampments['bin_hour'] = encampments.apply(bin_hours, axis=1)
encampments['new_date'] = encampments['Opened'].apply(lambda x:x[0:10])
encampments['Opened_rnd'] = encampments['new_date'] + ' '+ encampments['bin_hour'] + ':00:00'

encampments
```

```python
def make_poly(lon, lat):
    radius = Point(lon,lat).buffer(0.016)
    poly = Polygon(radius.exterior.coords)
    return poly

blocks['poly'] = blocks.apply(lambda row: make_poly(row['INTPTLON10'],row['INTPTLAT10']), axis=1)
```
